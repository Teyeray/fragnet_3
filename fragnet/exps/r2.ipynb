```python
import pickle
import numpy as np
from sklearn.metrics import r2_score

# 读取结果文件
with open('ft/ft_0331/test_res_42.pkl', 'rb') as f:
    res_test = pickle.load(f)

with open('ft/ft_0331/val_res_42.pkl', 'rb') as f:
    res_val = pickle.load(f)

true_test = np.array(res_test['true'])
pred_test = np.array(res_test['pred'])

true_val = np.array(res_val['true'])
pred_val = np.array(res_val['pred'])

# 计算 R²
r2_res = r2_score(true_test, pred_test)
r2_val = r2_score(true_val, pred_val)
print(f'Test R²: {r2_res:.4f}, Val R2 : {r2_val:.4f}')
```
