```python
from rdkit import Chem
import pandas as pd
```

```python
df = pd.read_csv('chemicalbook_pubchem_combined_filtered_smiles_new_old_pce.csv')

len(df)
```

```python
mask = df['SMILES'].str.startswith('.') | df['SMILES'].str.endswith('.')
if mask.any():
    print("以下行的SMILES头尾有'.'：")
    for idx, row in df[mask].iterrows():
        print(f"  第{idx}行: {row['SMILES']}")
    df['SMILES'] = df['SMILES'].str.strip('.')
    print(f"\n共处理 {mask.sum()} 条")
else:
    print("没有发现头尾有'.'的SMILES")
```

```python
df.to_csv('data_cleaned.csv', index=False)
```

```python
from rdkit import Chem
from tqdm import tqdm
def is_valid_smiles(smi):
    try:
        mol = Chem.MolFromSmiles(smi)
        return mol is not None
    except:
        return False

tqdm.pandas()
df["valid"] = df["SMILES"].progress_apply(is_valid_smiles)

invalid = df[~df["valid"]]
print(f"非法 SMILES 数量: {len(invalid)}")
print(f"合法 SMILES 数量: {df['valid'].sum()}")

# 保存非法的，方便排查
invalid[["SMILES"]].to_csv("invalid_smiles.csv", index=False)

# 保存过滤后的干净数据
df_clean = df[df["valid"]].drop(columns=["valid"]).reset_index(drop=True)

df_clean = df_clean.rename(columns={"SMILES":"smiles"})

df_clean.to_csv("book_cleaned.csv", index=False)
print("✅ 已保存 book_cleaned.csv")
```
